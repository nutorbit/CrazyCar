# Curve corner

If we know some x, y coordinate then we can calculate the angle field for following the circle curve.

Suppose the origin start with (0, 0)

$$
\theta = \tan^{-1}(\frac{x}{y})
$$

And, for any origin point (ox, oy)

$$
\theta = \tan^{-1}(\frac{x-ox}{y-oy})
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.zeros((100, 100))

for i in range(m.shape[0]):
    for j in range(m.shape[1]):
        m[i][j] = round(np.degrees(np.arctan2(i+1, j+1)))
        
m[::-1]
```

```python
plt.imshow(m[::-1])
```

```python
x,y = np.meshgrid(np.linspace(-5,5,100),np.linspace(-5,5,100))
```

```python
np.vectorize(round)(np.degrees(np.arctan2(x, y)))
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = -y/np.sqrt(x**2 + y**2)
v = x/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v)
plt.show()
```
